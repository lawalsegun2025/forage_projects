Currently we have 3 datasets. In order to include all of this data within a predictive model, we need to merge them together into 1 dataframe.

````python
from datetime import datetime

# Create a function that converts the `timestamp` column to hours
def convert_timestamp_to_hourly(data: pd.DataFrame=None, column: str=None):
  dummy = data.copy()
  new_ts = dummy[column].tolist()
  new_ts = [i.strftime('%Y-%m-%d %H:00:00') for i in new_ts]
  new_ts = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in new_ts]
  dummy[column] = new_ts
  return dummy
````

````python
sales_df = convert_timestamp_to_hourly(sales_df, 'timestamp')
sales_df.head()
````

````python
stock_df = convert_timestamp_to_hourly(stock_df, 'timestamp')
stock_df.head()
````

````python
temp_df = convert_timestamp_to_hourly(temp_df, 'timestamp')
temp_df.head()
````

Now we can see all the `timestamp` xolumns have had the minutes and seconds reduced to `00`. The next thinig to do, is to aggregate the datasets in order to combine rows which have the same value for `timestamp`.

Foe the `sales` data set, we want to group the data by `timestamp` but also by `product_id`. When we aggregate, we must choose which columns to aggregate by the grouping. For now, let's aggregate quantity.

````python
sales_df.head()
````

````python
# sales_agg = sales_df.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum', 
#                                                                'unit_price': 'mean'}).reset_index()
# sales_agg
````

````python
sales_agg = sales_df.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum', }).reset_index()
sales_agg.head()
````

We now have an aggregated sales data where each row represents a unique combination of hour during which the sales took place from that weeks worth of data and the product_id. We summed the quantity and we took the mean average of the unit_price. 

For the stock data, we want to group it the same way and agrregate the `estimated_stock_pct`

````python
stock_df.head()
````

````python
stock_agg = stock_df.groupby(['timestamp', 'product_id']).agg({'estimated_stock_pct': 'mean'}).reset_index()
stock_agg.head()
````

This shows us th average stock percentage of each product at unique hours within the week of asmple data.

Finally, for the temperature data, products_id does not exist in the table, so we simply need to group by timestamp and aggregate the `temperature`.

````python
temp_agg = temp_df.groupby(['timestamp']).agg({'temperature': 'mean'}).reset_index()
temp_agg.head()
````

This gives us the average temperature of the storage facility where the produce is stored in the warehouse by unique hours during th eweek. Now, we are ready to merge our data. We will use the `stock_agg` table as our base table, and we will merge our other 2 tables onto this.

````python
merged_df = stock_agg.merge(sales_agg, on=['timestamp', 'product_id'], how='left')
merged_df.head()
````

````python
merged_df = merged_df.merge(temp_agg, on=["timestamp"], how='left')
merged_df.head()
````
